### Passo 1 - Importar Arquivos e Bibliotecas

```python
# Para Ler as bases de dados
import pandas as pd
```

```python
# importar Bases de Dados
emails = pd.read_excel('Bases de Dados/Emails.xlsx')
 # , encoding="utf-8")  ou , encoding="latin1") 
lojas = pd.read_csv('Bases de Dados/Lojas.csv', encoding="ISO-8859-1", sep=';')
vendas = pd.read_excel('Bases de Dados/Vendas.xlsx')

display(emails)
display(lojas)
display(vendas)
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
# Incluir a Coluna 'Nome Da Loja' na Planilha 'Vendas.xlsx'
# 'on="ID Lojas"' -> É a coluna em Comum nas duas Planilhas
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python
# TINHA FEITO MERGE DUAS VEZES, GERANDO DUAS COLUNAS REPETIDAS. APAGUEI 
#..COM DROP:
vendas.drop("Loja_x", axis=1, inplace=True)
vendas.drop("Loja_y", axis=1, inplace=True)
```

```python
# Percorre as lojas da planilha Lojas em coluna Lojas (desta ultima)
dicionario_de_lojas = {}

    # A cada Loop detecta uma loja, Cria uma chave com seu nome
    #..Quando -> Vendas\Coluna -> Loja for igual ao que esta 
    #..sendo Percorrido no Loop atual
    
    #Dicionario recebe Valor que esta sendo percorrido, quando este
    #..for igual ao que esta na coluna Loja da planilha 'Vendas'

# Lojas -> Contém o CSV com o nome das lojas:
# lojas['Loja'] -> Pegaa a coluna ['Loja'] do CSV lojas e insere no
#..dicionario 'dicionario_de_lojas' (usando o sinal = )
for loja in lojas['Loja']:
    dicionario_de_lojas[loja] = vendas.loc[vendas['Loja'] == loja, :]

                                      
display(dicionario_de_lojas['Shopping Vila Velha'])                                           
display(dicionario_de_lojas['Salvador Shopping'])

```

```python
# Mostrar Maior/Data_Mais_Recente da coluna:
dia_indicador = vendas['Data'].max()
print(dia_indicador)

print(f'{dia_indicador.day}/{dia_indicador.month}/{dia_indicador.year}')

# Se a tabela receber um valor/Data maior, isso sera atualizado automaticamente.
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
import pathlib

caminho_do_Backup = pathlib.Path(r'Backup Arquivos Lojas/')

# Transforma um diretorio num iteravel, e torna possivel percorrer &
#.. listar todas as pastas dentro dele.
arquivos_pasta_backup = caminho_do_Backup.iterdir()


# arquivos_pasta_backup -> Contém o Caminho com as
#.. pastas/nomes_de_todos_os_Shoppings :
lista_nomes_backup = []
for r in arquivos_pasta_backup:
    #print(r)
    #Pega só o nome da(s) pasta(s)
    #print(r.name)
    
    # Criamos uma lista com o nome de tudo o que estiver dentro da pasta
    lista_nomes_backup.append(r.name)

#print(lista_nomes_backup)


if 'Shopping Eldorado' in lista_nomes_backup:
    print('Its Here')
```

```python
# Agora o mesmo procedimento, porém com ListComprehension:

# Se esta linha Não estiver na mesma Célula -> Não funciona:
arquivos_pasta_backup = caminho_do_Backup.iterdir()

lista_nomes_backup = [r.name for r in arquivos_pasta_backup]

# Nome de todas as pastas dentro da pasta (Lista_Nomes_Backuo):
print(lista_nomes_backup)

# for r in arquivos_pasta_backup:
#     print(r.name)
```

```python
# Percorre o Dicionario que contem as informações unicas de cada Loja
for loja in dicionario_de_lojas:
    # Se esta loja não estiver na lista_nomes_backup (Que é a pasta que
    #..contém as Pastas com o Nome de todas as lojas)
    if loja not in lista_nomes_backup:
        # caminho_do_Backup -> Possui o caminho (Pathlib) Para a Pasta Backup
        #..loja é o nome da chave ou lojas no Dicionário
        
        nova_pasta = caminho_do_Backup / loja
        # Será criada uma pasta para cada nome/chave -> 'loja'
        nova_pasta.mkdir()
    
    # Dia Indicador é -> dia_indicador = vendas['Data']   , a Data do DF
    nome_arquivo = f'{dia_indicador.month}_{dia_indicador.day}_{loja}.xlsx'
    #nome_arquivo = '{}_{}_{}.xlsx'.format(dia_indicador.month, dia_indicador.day, loja)
    
    # caminho do arquivo ref ->'Backup Arquivos Lojas/'
    # 'loja' -> Item percorrido da pasta -> Backup Arquivos Lojas/
    # 'nome_arquivo' -> Nome gerado a partir de informações de data dos arquivos.
    local_arquivo = caminho_do_Backup / loja / nome_arquivo
    
    
    
    print(nome_arquivo)
    
    # Transforme Tal Dataframe em xlsx & coloque no 'local_arquivo'
    dicionario_de_lojas[loja].to_excel(local_arquivo)
    
```

### Passo 4 - Calcular o indicador para 1 loja

```python
# Variavel...
loja = 'Norte Shopping'

# INDICADOR 1 - FATURAMENTO
vendas_loja = dicionario_de_lojas[loja]

# dia_indicador -> Contém a ultima OU a maior (.max) data do XLSX
vendas_loja_dia = vendas_loja.loc[vendas_loja['Data']==dia_indicador, :]

# Faturamento
faturamento_ano =vendas_loja['Valor Final'].sum()
# Soma da coluna 'Valor Final'
print(faturamento_ano)

faturamento_do_dia = vendas_loja_dia['Valor Final'].sum()
#É a soma das vendas deste ultimo dia. Ou Qualquer valor de venda em
#..uma linha que referencie o 'Norte Shopping' neste ultimo dia.
print(faturamento_do_dia) 

# INDICADOR 2 - DIVERSIDADE DE PRODUTOS

# DIVERSIDADE DE PRODUTOS POR ANO:
# Filtrar -> Apenas valores Unicos aparecerão
# print(vendas_loja['Produto'].unique()) -> Quantos tipos de produtos existem nesta loja
qtd_prod_ano = len(vendas_loja['Produto'].unique())
print(qtd_prod_ano) #Quant de prod vendidos no ano pela 'Norte Shopping'

# DIVERSIDADE DE PRODUTOS POR DIA:
# No ultimo dia (dia_indicador) -> Esta foi a quantidade de produtos vendida:
qtd_prod_dia = len(vendas_loja_dia['Produto'].unique())
print(qtd_prod_dia)
#__________________________________
# INDICADOR 3 - TICKET MÉDIO:

# Ticket médio
# Somou o valor de todas as colunas que tenham o
#..menos a coluna 'Código Venda'
#..Todas as colunas agora aparecem com a soma Delas mesmas E por isso
#..em menos linhas:
valor_venda = vendas_loja.groupby('Código Venda').sum()
display(valor_venda)

# MÉDIA:
media_valor_venda = valor_venda['Valor Final'].mean()
display(media_valor_venda)

# TICKET MÉDIO DIA
valor_venda_dia = vendas_loja_dia.groupby('Código Venda').sum()
display(valor_venda_dia)

# A média do valor final da Linha de Cima (que foi Um resultado Só == 1259  :
ticket_medio_dia = valor_venda_dia['Valor Final'].mean()
print(ticket_medio_dia)
```
